# 1. LDA for Topic Modeling in Python


In this section we will see how Python can be used to implement LDA for topic
modeling.
The data set contains user reviews for different products in the food category. We
will use LDA to group the user reviews into 5 categories.
The first step, as always, is to import the data set along with the required libraries.
Execute the following script to do so:

```python
import random
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import random
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.decomposition import NMF
print(os.listdir("../input"))
```

```python

reviews_datasets = pd.read_csv('../input/amazon-fine-food-reviews/Reviews.csv')
reviews_datasets = reviews_datasets.head(20000)
reviews_datasets.dropna()
```

```python
#count vect for LDA
count_vect = CountVectorizer()
doc_term_matrix = count_vect.fit_transform(reviews_datasets['Text'])
```

```python
LDA = LatentDirichletAllocation(n_components=5, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
for i in range(10):
    random_id = random.randint(0,len(count_vect.get_feature_names()))
    print(count_vect.get_feature_names()[random_id])
```

```python
first_topic = LDA.components_[0]
```

```python
top_topic_words = first_topic.argsort()[-10:]
```

```python
for i in top_topic_words:
    print(count_vect.get_feature_names()[i])
```

```python
for i,topic in enumerate(LDA.components_):
    print(f'Top 10 words for topic #{i}:')
    print([count_vect.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
topic_values = LDA.transform(doc_term_matrix)
topic_values.shape
```

```python
reviews_datasets['Topic_LDA'] = topic_values.argmax(axis=1)
```

```python
reviews_datasets.head(10)
```

```python
reviews_datasets.Topic_LDA.value_counts()
```

# 2. NMF for Topic Modeling in Python

the last part. You will see that the steps are also quite similar.

```python
tfidf_vect = TfidfVectorizer()
doc_term_matrix = tfidf_vect.fit_transform(reviews_datasets['Text'])
```
